3-2: Maximum Value of the Loot (Fractional Knapsack)

```python
n, W = [int(i) for i in input().split()]
lst = []

if W == 0:
    print(0)
    quit()

for _ in range(n):
    v, w = [int(i) for i in input().split()]
    if v == 0:
        continue
    lst.append((v, w))

lst.sort(key = lambda x: x[0]/x[1], reverse = True)

total_value = 0

for v,w in lst:
    if W==0:
        print(total_value)
        quit()
    amt = min(w, W)
    total_value += amt*v/w
    W -= amt

print(total_value)
```

3-3: Car Fueling

```python
def compute_min_refills(distance, tank, stops):
    if distance <= tank:
        return 0
    i = 0
    num_stops = 0
    current_stop = 0
    dist_covered = 0
    flag = False
    stops.append(distance)
    while dist_covered < distance:
        while i < len(stops) and stops[i] - current_stop <= tank:
            i += 1
            flag = True
        if flag:
            if i == len(stops):
                break
            current_stop = stops[i-1]
            dist_covered = current_stop
            num_stops +=1
        else:
            return -1
        flag = False
    return num_stops

d = int(input())
m = int(input())
l = int(input())
stops = list(map(int,input().split()))
print(compute_min_refills(d, m, stops))
```

3-5: Collecting Signatures (Covering Segments by Points)

```python
n = int(input())
lst = []

for _ in range(n):
    a, b = [int(i) for i in input().split()]
    lst.append((a,b))

lst.sort(key = lambda x: x[1])

index = 0
coordinates = []

while index < n:
    curr = lst[index]
    while index < n-1 and curr[1]>=lst[index+1][0]:
        index += 1
    coordinates.append(curr[1])
    index += 1
print(len(coordinates))
print(' '.join([str(i) for i in coordinates]))
```

3-6: Maximum Number of Prizes (Different Summands)

```python
n = int(input())
if n == 1:
    print(1)
    print(1)
    quit()
W = n
prizes = []
for i in range(1, n):
    if W>2*i:
        prizes.append(i)
        W -= i
    else:
        prizes.append(W)
        break

print(len(prizes))
print(' '.join([str(i) for i in prizes]))
```

3-7: Maximum Salary (Largest Number)

```python
n = int(input())
lst = [int(i) for i in input().split()]


def IsGreaterOrEqual(digit, max_digit):
    return int(str(digit)+str(max_digit))>=int(str(max_digit)+str(digit))

def largestnumber(lst):
    answer = []
    
    while lst!=[]:
        max_digit = 0
        for digit in lst:
            if IsGreaterOrEqual(digit, max_digit):
                max_digit = digit
        answer.append(max_digit)
        lst.remove(max_digit)

    return answer

print(''.join([str(i) for i in largestnumber(lst)]))
```
